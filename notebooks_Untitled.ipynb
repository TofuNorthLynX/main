```python
# Import the necessary modules

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data = pd.read_csv('mcs_ds_edited_iter_shuffled.csv')
  
df = pd.DataFrame(data)
  
X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 4])
  
# Plot the data using bar() method
plt.bar(X, Y)
plt.title("trans_range")
  
# Show the plot
plt.show()
```
